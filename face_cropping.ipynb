```python
import image_processing
```

```python
angry = image_processing.read_images('dataset/train/angry/*.*')
contempt = image_processing.read_images('dataset/train/contempt/*.*')
fear = image_processing.read_images('dataset/train/fear/*.*')
happy = image_processing.read_images('dataset/train/happy/*.*')
neutral = image_processing.read_images('dataset/train/neutral/*.*')
sad = image_processing.read_images('dataset/train/sad/*.*')
surprise = image_processing.read_images('dataset/train/surprise/*.*')
```

```python
angry = [img for img in angry if img is not None]
contempt = [img for img in contempt if img is not None]
fear = [img for img in fear if img is not None]
happy = [img for img in happy if img is not None]
neutral = [img for img in neutral if img is not None]
sad = [img for img in sad if img is not None]
surprise = [img for img in surprise if img is not None]
```

```python
emotions = [angry, contempt, fear, happy, neutral, sad, surprise]
```

```python
for i, emotion in enumerate(emotions):
    print(f'{i} length is {len(emotion)}')
```
